Q 5.Why and When to use While loop in python give a detailed description with example

Answer:-

When to Use a While Loop in Python

A while loop is used when you want to repeat a block of code as long as a certain condition is true. It is particularly useful in situations where the number of iterations is not known beforehand.

Common Use Cases

User Input Validation: Keep asking for input until the user provides valid data.

Dynamic Conditions: Continue processing until a specific condition changes (e.g., waiting for a signal).

Infinite Loops: Create loops that run indefinitely until a break condition is met.


Example

Here’s a simple example that prompts the user for a positive number:

```python
number = -1  # Initialize with an invalid number

while number <= 0:
    number = int(input("Enter a positive number: "))
    if number <= 0:
        print("That's not positive. Try again.")

print(f"You entered: {number}")
```

Explanation

The loop continues until the user enters a positive number.

If the input is not valid, it prompts the user again.

Conclusion

Use a while loop when the number of iterations is uncertain and depends on dynamic conditions. Just ensure that the loop will eventually terminate to avoid infinite loops.



Q 5.Use nested while loop to print 3 different patterns

Answer:-

```python
# Pattern 1: Right Angled Star pattern
i = 1
while i <= 10:
    j = 1
    while j <= i:
        print("*", end="")
        j += 1
    print("")
    i += 1
```

```python
# Pattern 2: Isoceles Triangle Star pattern
n = 10
i = 1
while i <= n:
    spaces = n - i
    while spaces > 0:
        print(" ", end="")
        spaces -= 1
    stars = 2 * i - 1
    while stars > 0:
        print("*", end="")
        stars -= 1
    print("")
    i += 1
```

```python
#Pattern 3: Diamond
n = 5
i = 1
while i <= n:
    spaces = n - i
    while spaces > 0:
        print(" ", end="")
        spaces -= 1
    stars = 2 * i - 1
    while stars > 0:
        print("*", end="")
        stars -= 1
    print("")
    i += 1

i = n - 1
while i >= 1:
    spaces = n - i
    while spaces > 0:
        print(" ", end="")
        spaces -= 1
    stars = 2 * i - 1
    while stars > 0:
        print("*", end="")
        stars -= 1
    print("")
    i -= 1
```
